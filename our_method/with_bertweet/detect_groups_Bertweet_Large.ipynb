```python
from sklearn.utils import resample

# Separate majority and minority classes
df_majority = G1_data[G1_data['G1_label'] == 0]
df_minority = G1_data[G1_data['G1_label'] == 1]

# Downsample the majority class
df_majority_downsampled = resample(df_majority,
                                   replace=False,  # Sample without replacement
                                   n_samples=len(df_minority)*4,  # Match minority class size
                                   random_state=42)  # For reproducibility

G1_data = pd.concat([df_majority_downsampled, df_minority])
df1 = G1_data[['text', 'G1_label' ]]
df1= df1.rename(columns={'G1_label': 'label'})
df1
```

```python
from simpletransformers.classification import ClassificationModel, ClassificationArgs

from sklearn.metrics import f1_score, recall_score, precision_score

def f1_multiclass(labels, preds):
    return f1_score(labels, preds, average='binary')

def prec_multiclass(labels, preds):
    return precision_score(labels, preds, average='binary')

def recall_multiclass(labels, preds):
    return recall_score(labels, preds, average='binary')

model = ClassificationModel('roberta', 'vinai/bertweet-large',
                            weight=[0.5,2],
                                      args={'train_batch_size':4,
                                            "evaluate_during_training":True,'overwrite_output_dir': True,'gradient_accumulation_steps':16, 'learning_rate': 3e-5, 'num_train_epochs': 8, 'max_seq_length': 256, 'threashold':0.5})
```
