```python
from datasets import load_dataset

dataset = load_dataset("csv", data_files={'train': "/content/train_3_zero.csv", 'test':"/content/test_3_zero.csv"})
```

```python
dataset
```

```python
def encode_labels(record):
    return {"labels": [record[feature] for feature in [ 'Repetition', 'Doubt', 'noPropaganda']]}

dataset = dataset.map(encode_labels)
dataset
```

```python
from setfit import SetFitModel
from sentence_transformers import SentenceTransformer
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression
# model = SetFitModel.from_pretrained('sentence-transformers/paraphrase-mpnet-base-v2', multi_target_strategy="one-vs-rest")

model = SetFitModel(
                    model_body=SentenceTransformer('sentence-transformers/paraphrase-mpnet-base-v2'),
                    model_head=OneVsRestClassifier(LogisticRegression(class_weight="balanced")),
                    )
```

```python
model.to("cuda")
```

```python
from setfit import SetFitModel, Trainer, TrainingArguments, sample_dataset

args = TrainingArguments(
    batch_size=16,
    num_epochs=2,
    evaluation_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,
    num_iterations=5,
)

trainer = Trainer(
    model=model,
    args=args,
    train_dataset=dataset['train'],
    eval_dataset=dataset['test'],
    column_mapping={"text": "text", "target": "label"}  # Map dataset columns to text/label expected by trainer
)
```

```python
# Train and evaluate
trainer.train()
```

```python
metrics = trainer.evaluate(dataset['test'])
print(metrics)
```
