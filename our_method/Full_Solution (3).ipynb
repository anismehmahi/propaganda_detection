```python
import numpy as np
import pandas as pd
```

```python
def predict_row(row):
    g1_g3 = [model.predict(row['text']) for model in [g1, g3]]
    g2_preds = g2(row['text']).detach().cpu().numpy().tolist()

    g2_rep_dout_pred = g2_rep_dout(row['text']).item()
    #print(g2_rep_dout_pred)
    g2_preds[1]=g2_preds[4]=0
    if g2_rep_dout_pred==0: #repe
      g2_preds[1]=1
    elif g2_rep_dout_pred==1: # doubt
      g2_preds[4]=1

    # g2_preds[1]=g2_rep_dout_preds.item()# repe
    # g2_preds[4]=g2_rep_dout_preds[1] # doubt
    # g2_preds[-1]=g2_rep_dout_preds[2] # no propaganda
    # res = int(g2_rep_dout_preds[0]['label'][-1]) # 0 means repetition or 1 means doubt or 2 means not propaganda
    # if res==0:
    #   g2_preds[1] = 1
    # elif res==1:
    #   g2_preds[4] = 1

    predictions = g1_g3 + g2_preds[:-1]
    no_propa = g2_preds[-1]
    if any(predictions) & (no_propa==0):
        return 'P'
    elif not any(predictions) & (no_propa==0):
        return 'O'
    elif not any(predictions) & (no_propa==1):
        return 'O'
    else: # 1 and 1
        return 'O'
```

```python
test_multi = pd.read_csv("/content/test-multi-label-oana.csv")
test_multi['label']= 'P'
nonPropa = pd.read_csv("/content/final-test.csv")
nonPropa = nonPropa[nonPropa['Technique']=='Not propaganda'][['text', 'label']]
```

```python
final_test_with_no_propaganda = pd.concat([test_multi[['text', 'label']], nonPropa[['text', 'label']]], ignore_index =True)
final_test_with_no_propaganda.to_csv("final_test_with_no_propaganda.csv", index = False)
```

```python
final_test_with_no_propaganda['y_pred'] = final_test_with_no_propaganda.apply(predict_row, axis=1)
```

```python
from sklearn.metrics import classification_report
print(classification_report(final_test_with_no_propaganda['label'] ,final_test_with_no_propaganda['y_pred'] ))
```
